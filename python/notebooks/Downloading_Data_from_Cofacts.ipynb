## Retrieving Individual Tables

The Cofacts dataset is organized into [normalized tables](https://huggingface.co/datasets/Cofacts/line-msg-fact-check-tw#data-subsets-and-their-fields). To access a specific table, you can utilize the `load_dataset` function with the argument `"Cofacts/line-msg-fact-check-tw"`. This allows you to retrieve the desired table for your analysis and research purposes.

```python
#@title `load_dataset` loads articles by default
articles = pd.DataFrame(load_dataset("Cofacts/line-msg-fact-check-tw")["train"])
```

```python
articles.to_csv("../../data/articles.csv", index=False)
```
