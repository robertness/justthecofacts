## Retrieving Individual Tables

The Cofacts dataset is organized into [normalized tables](https://huggingface.co/datasets/Cofacts/line-msg-fact-check-tw#data-subsets-and-their-fields). To access a specific table, you can utilize the `load_dataset` function with the argument `"Cofacts/line-msg-fact-check-tw"`. This allows you to retrieve the desired table for your analysis and research purposes.

```python
#@title `load_dataset` loads articles by default

articles = load_dataset("Cofacts/line-msg-fact-check-tw")
articles
```

```python
#@title Specify table

article_replies = load_dataset("Cofacts/line-msg-fact-check-tw", "article_replies")
article_replies
```

## Combining Multiple Tables to Form a Dataset

The Cofacts dataset comprises [normalized tables](https://huggingface.co/datasets/Cofacts/line-msg-fact-check-tw#tables-and-their-fields), and you have the flexibility to merge and match these tables to create a customized dataset tailored to your specific needs.

### Use Case: Rumor / Non-rumor / Opinion Classifier

For instance, if your objective is to build a classifier for rumor detection, you will require a dataset that includes text and corresponding rumor labels. These pieces of information can be sourced from two distinct tables:

- The `text` field can be found in the [table `articles`](https://huggingface.co/datasets/Cofacts/line-msg-fact-check-tw#articles).
- The `replyType` field, which contains crowd-sourced labels such as `RUMOR` (indicating the presence of disinformation) or `NOT_RUMOR` (indicating the presence of accurate information), is available in [table `article_replies`](https://huggingface.co/datasets/Cofacts/line-msg-fact-check-tw#article_replies).

To merge these two tables and obtain a unified dataset, you can follow these steps:

1. Load the `articles` and `article_replies` tables separately using the `load_dataset()` function.
2. Utilize the `pandas` library to perform a join operation between these two tables, enabling you to consolidate the `text` and `replyType` fields into a single table.

![ER model of articles and article_replies](https://docs.google.com/drawings/d/e/2PACX-1vT_2eSpDmj3ah-BRu1jdElupIqJUaP_fjzjc6LTE4YO8yPF-exiM8IFbfPp1c8QHBbUhx_MZheTE4om/pub?w=578&h=221)

```python
import pandas as pd
article_with_replies = pd.merge(articles['train'].to_pandas(), article_replies['train'].to_pandas(), left_on='id', right_on='articleId', how='left')

article_with_replies
```

Once you have loaded the `articles` and `article_replies` tables and joined them to obtain the `text` and `replyType` fields together, you can extract these fields to form your training data.

```python
classifier_traindata = article_with_replies[['text', 'replyType']]
classifier_traindata
```

### Case: Fine-Tuning LLM for Internet Message Responses

To fine-tune a Large Language Model (LLM) for automatically generating responses to internet messages, you'll need a dataset containing internet messages and their corresponding reply text. These pieces of text can be obtained from two distinct tables within the Cofacts dataset:

- The `text` field in the [table `articles`](https://huggingface.co/datasets/Cofacts/line-msg-fact-check-tw#articles): This field contains messages that have circulated in LINE and have been reported by Cofacts chatbot users.
- The `text` field in [table `replies`](https://huggingface.co/datasets/Cofacts/line-msg-fact-check-tw#replies): This table contains crowd-sourced fact-checking replies composed by community fact checkers.

It's important to note that there is a many-to-many relationship (N-N mapping) between `articles` and `replies`. This means that one article can have multiple replies, and one reply can be used in response to multiple articles. The relationship between these tables is recorded in the linking table `articleReplies`.

![ER model of articles, article_replies, and replies](https://docs.google.com/drawings/d/e/2PACX-1vS-w-ZdlfX8LNrQjsc-7XwelpHBexDnAA_F03OJsINQlDce8ZNmke0kxgq29eVgyvgPD-EA7b2ifmgX/pub?w=906&h=226)

1. Load the `articles`, `article_replies`, and `replies` tables separately using the `load_dataset()` function.
2. Utilize the `pandas` library, along with appropriate join operations, to merge these three tables together. This will enable you to obtain the `text` from both the `article` and the `reply` in a single table.

Here is a code snippet illustrating how you can achieve this:

```python
replies = load_dataset("Cofacts/line-msg-fact-check-tw", "replies")


article_and_replies = pd.merge(
    pd.merge( articles['train'].to_pandas(), article_replies['train'].to_pandas(), left_on='id', right_on='articleId', how='inner' ),
    replies['train'].to_pandas(), left_on='replyId', right_on='id', how='inner'
)

article_and_replies
```

You can use the `text_x` and `text_y` columns from the `article_and_replies` DataFrame to create prompts for fine-tuning your Language Model (LLM). These columns can serve as input-output pairs for training your model.


```python
finetune_corpus = article_and_replies[['text_x', 'text_y']]
finetune_corpus
```
